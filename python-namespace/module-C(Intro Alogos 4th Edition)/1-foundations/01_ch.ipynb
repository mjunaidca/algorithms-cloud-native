### 1.1-1
**Real-world example requiring sorting:**
Imagine you're managing a library, and new books arrive daily. To make it easier for patrons to find the books they need, you must arrange them in alphabetical order by author last name or title. This sorting process ensures that books are quickly and efficiently located.

**Real-world example requiring finding the shortest distance between two points:**
Consider a delivery driver who needs to drop off packages at multiple locations in a city. The driver wants to find the shortest possible route from the warehouse to each delivery point to minimize travel time and fuel consumption.

### 1.1-2
**Other measures of efficiency to consider:**
- **Memory Usage:** How much memory does the algorithm consume while executing? This is crucial for devices with limited memory resources.
- **Scalability:** How well does the algorithm perform as the size of the input data increases? This is important for applications that deal with large datasets.
- **Energy Consumption:** Especially in mobile or embedded systems, the amount of energy consumed by the algorithm can be a critical factor.
- **Robustness:** The ability of the algorithm to handle unexpected input or scenarios without failing.
- **Ease of Implementation and Maintenance:** How easy is it to implement, understand, and maintain the algorithm? This can affect the long-term viability of a solution.

### 1.1-3
**Data Structure: Hash Table**
- **Strengths:**
  - **Fast Access:** Provides average-case O(1) time complexity for search, insert, and delete operations, making it very efficient for these tasks.
  - **Ease of Use:** Simple to implement for basic key-value storage needs.
- **Limitations:**
  - **Collisions:** Although collisions (where two keys hash to the same index) can be managed with techniques like chaining or open addressing, they can degrade performance.
  - **Memory Usage:** Hash tables can require more memory than other data structures, especially if the load factor is kept low to minimize collisions.
  - **Ordering:** Hash tables do not maintain any order of elements, so if ordered data retrieval is required, another data structure might be more appropriate.

### 1.1-4
**Similarities between the shortest-path and traveling-salesperson problems:**
- Both problems involve finding an optimal path in a graph.
- They require consideration of distances or costs between points and aim to minimize the total distance or cost.

**Differences:**
- The **shortest-path problem** seeks the shortest path between two specific points. It focuses on finding the most efficient way to get from point A to point B.
- The **traveling-salesperson problem (TSP)** requires finding the shortest possible route that visits every point exactly once and returns to the starting point. TSP is more complex and involves considering the entire set of points.

### 1.1-5
**Problem requiring the best solution:**
- **Example:** Designing a pacemaker. The device must function correctly at all times because any failure could have serious or fatal consequences. Only the best, most reliable solution is acceptable.

**Problem where an approximately best solution is good enough:**
- **Example:** Streaming video quality. In many cases, a high-definition (HD) video is good enough for viewers. While 4K might be better, it's not necessary for everyone, and a slightly lower resolution will still provide a satisfactory experience.

### 1.1-6
**Real-world problem with varying input availability:**
- **Example:** Stock market analysis. Sometimes, you might have all the historical data needed to make an investment decision in advance. However, other times, new information about market conditions, company earnings, or economic events arrives continuously throughout the day. You must update your strategy in real-time as this new data becomes available.
